# 5 Extract a set of normalized tables. That is, each entry of a normalized table must contain exactly one value (not a list or a dictionary). There is no need to use SQL for this point.

```python
def json_to_tables(df,json_column,id_column):
    aux_acu=pd.DataFrame()
    for i in range(len(df)):
        aux = pd.read_json(df[json_column][i])
        aux['movies_id'] = df[id_column][i]
        aux_acu=pd.concat([aux_acu,aux])
    return aux_acu
```

# 10 For each company and each decade, compute the overall revenue

```python
movies_revenue_per_company = pd.merge(movies[['id','original_title','revenue']],
                                     movies_production_companies.groupby('movies_id')[['name']].count(),
                                     left_on='id',
                                     right_index = True,
                                     how='right')
```

```python
movies_revenue_per_company.head(3)
```

```python
movies_revenue_per_company['revenue_divided']=movies_revenue_per_company['revenue']/movies_revenue_per_company['name']
movies['revenue_divided']= movies_revenue_per_company['revenue_divided']
movies[['revenue','id','production_companies','revenue_divided']].head(3)
```

```python
movies_companies_parameters = pd.merge(movies[['id','original_title','year','revenue','release_date']],
                                     movies_production_companies,
                                     left_on='id',
                                     right_on = 'movies_id',
                                     how='right')
```

```python
movies_companies_parameters.head(3)
```

```python
movies_companies_parameters.groupby(['name',pd.cut(movies_companies_parameters['year'],
      np.arange(movies_companies_parameters['year'].min(),
          movies_companies_parameters['year'].max()+1,
          10))]).agg({'revenue':'sum'}).head(3)
```

this is for divided revenue on the amouth of companies

```python
movies_companies_parameters_divided = pd.merge(movies[['id','original_title','year','revenue_divided','release_date']],
                                     movies_production_companies,
                                     left_on='id',
                                     right_on = 'movies_id',
                                     how='right')
```

```python
movies_companies_parameters.head(3)
```

```python
movies_companies_parameters_divided.groupby(['name',pd.cut(movies_companies_parameters_divided['year'],
      np.arange(movies_companies_parameters_divided['year'].min(),
          movies_companies_parameters_divided['year'].max()+1,
          10))]).agg({'revenue_divided':'sum'}).head(3)
```

# 11 For each decade, compute the company with maximum revenue

```python
idx = movies_companies_parameters.groupby(pd.cut(movies_companies_parameters['year'],
      np.arange(movies_companies_parameters['year'].min(),
          movies_companies_parameters['year'].max()+1,
          10)))['revenue'].transform(max) == movies_companies_parameters['revenue']
movies_companies_parameters[idx].sort_values('year')
```

```python
idx = movies_companies_parameters_divided.groupby(pd.cut(movies_companies_parameters_divided['year'],
      np.arange(movies_companies_parameters_divided['year'].min(),
          movies_companies_parameters_divided['year'].max()+1,
          10)))['revenue_divided'].transform(max) == movies_companies_parameters_divided['revenue_divided']
movies_companies_parameters_divided[idx].sort_values('year')
```
